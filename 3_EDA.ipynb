## **ANOVA - ANalysis Of VAriation**

As we said before ANOVA is a statistical method that stands for "Analysis of variations".<br>

This method can be used to find correlation between different groups of variables.<br>

We obtain from this method two values:


1.   F-test score: the ratio of variation between the groups's mean over the variation the ratio of variation between the groups's mean over the variation. https://en.wikipedia.org/wiki/F-test
2.   p-value: compute the statistical significance

To take in mind we have if the F-test score is less than 1 and the p-value is larger than 0.05 the data is not significantly different, but if the F-test score is higher than one and the p-value is larger than 0.05 then both variables are significantly different.

According to that we can said that there exists a strong correlation between variables if the ANOVA test give us a larger F-value score and a small p-value (in some cases we take values less than 0.05 and if we want more precision les than 0.01, the last one use with medical information) 

## **Correlation**

Correlation in statistics it normally refers to the degree to which a pair of variables are linearly related (interdependant). So we have to see when one variable increase the other could mantain the same behaviour or also could be the opposite.
To take into account this mean not is the same as causation.
We have positive correlation, negative correlation and weak correlation, this could be seen in the scatter plot descriptions.

**Pearson correlation**
We can measure the strenght of the correlation, so the pearson correlation give us two values:


1.   Correlation coefficient:

*   Large positive relation values close to +1.
*   Large negative relation values close to-1.
*   No relationship values close to 0.


2.   P-value:
*   Strong certainty in the result with P-value < 0.001
*   Moderate certainty in the result with P-value < 0.05
*   Weak certainty in the result with P-value < 0.1
*   No certainty in the result with P-value > 0.1
